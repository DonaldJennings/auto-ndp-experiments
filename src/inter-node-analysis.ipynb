```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read in the data
faasm_default = pd.read_csv('../Data/Dissertation Data/inter-node policies/faasm_default.csv')
most_slots = pd.read_csv('../Data/Dissertation Data/inter-node policies/most_slots.csv')
least_loaded = pd.read_csv('../Data/Dissertation Data/inter-node policies/least_load.csv')

print("Median Faasm Default: ", faasm_default['Latency'].median())
print("Median Most Slots: ", most_slots['Latency'].median())
print("Median Least Loaded: ", least_loaded['Latency'].median())
```

```python
# Get statistics
faasm_default_stats = faasm_default.describe()
most_slots_stats = most_slots.describe()
least_loaded_stats = least_loaded.describe()

print("Faasm Default: ", faasm_default_stats, "\n\n")



```

```python
print("Most Slots: ", most_slots_stats, "\n\n")
```

```python
print("Least Loaded: ", least_loaded_stats, "\n\n")
```

```python
# plot the data
fig, ax = plt.subplots()
faasm_default['Latency'].plot.kde(ax=ax, legend=True, title='Faasm Default vs. Most Slots vs. Least Loaded', label='Faasm Default')
most_slots['Latency'].plot.kde(ax=ax, legend=True, label='Most Slots')
least_loaded['Latency'].plot.kde(ax=ax, legend=True, label='Least Loaded')
ax.grid(axis='x')
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.grid(axis='y')
ax.set_xlabel('Latency (s)')
ax.set_ylabel('Density')
plt.show()
```
