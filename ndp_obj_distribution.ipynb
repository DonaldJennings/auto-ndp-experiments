```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the data
single_obj_data = pd.read_csv('./25g data/ndp - single obj vs multiple obj/single_obj_ndp_call_faasm_default_iters_250.csv')
multiple_obj_data = pd.read_csv('./25g data/ndp - single obj vs multiple obj/multiple_objs_ndp_call_faasm_default_iters_250.csv')
```

```python
# Print mean latency
print("Single obj mean latency: ", single_obj_data['Latency'].mean())
print("Multiple obj mean latency: ", multiple_obj_data['Latency'].mean())

```

```python
# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(single_obj_data['Latency'], label='Single Object')
plt.plot(multiple_obj_data['Latency'], label='Multiple Objects')
plt.xlabel('Invocation')
plt.ylabel('Latency (ms)')
plt.title('NDP Call Latency - Faasm Default Invoker')
plt.legend()
plt.show()
```

```python
rr_single_obj_data = pd.read_csv('./25g data/ndp - single obj vs multiple obj/single_obj_ndp_call_rr_iters_250.csv')
rr_multiple_obj_data = pd.read_csv('./25g data/ndp - single obj vs multiple obj/multiple_objs_ndp_call_rr_iters_250.csv')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(rr_single_obj_data['Latency'], label='Single Object')
plt.plot(rr_multiple_obj_data['Latency'], label='Multiple Objects')
plt.xlabel('Invocation')
plt.ylabel('Latency (ms)')
plt.title('NDP Call Latency - Faasm Round Robin Invoker')
plt.legend()
plt.show()
```

```python
# Plot 2 graphs on same figure
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
ax[0].plot(single_obj_data['Latency'], label='Single Object')
ax[0].plot(multiple_obj_data['Latency'], label='Multiple Objects', alpha=0.6)
ax[0].set_xlabel('Invocation')
ax[0].set_ylabel('Latency (s)')
ax[0].set_title('NDP Call Latency - Faasm Default Invoker')
ax[0].legend()

ax[1].plot(rr_single_obj_data['Latency'], label='Single Object')
ax[1].plot(rr_multiple_obj_data['Latency'], label='Multiple Objects', alpha=0.6)
ax[1].set_xlabel('Invocation')
ax[1].set_ylabel('Latency (s)')
ax[1].set_title('NDP Call Latency - Multi-tiered invoker')
ax[1].legend()

plt.show()
```

```python
# print median latency for both invokers
print("Single obj median latency - default: ", round(single_obj_data['Latency'].median(),2))
print("Multiple obj median latency - default: ", round(multiple_obj_data['Latency'].median(),2))
print("Single obj median latency - multi_tiered: ", round(rr_single_obj_data['Latency'].median(),2))
print("Multiple obj median latency - multi_tiered: ", round(rr_multiple_obj_data['Latency'].median(),2))
```
